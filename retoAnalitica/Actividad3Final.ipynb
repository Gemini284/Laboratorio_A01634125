```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import random
import seaborn as sns; sns.set()
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Gemini284/Laboratorio_A01634125/main/retoAnalitica/delitos-por-feminicidio-en-jalisco-por-municipio-periodo-2016-agosto2021.csv')
df.head()
```

```python
fig, axs = plt.subplots(figsize=(4,4))

bplot = sns.boxplot(data = df, y = '2018', x = '2017')

plt.suptitle('delitos por municipio en el a単o 2018',y=1.05)
```

```python
fig, axs = plt.subplots(figsize=(4,4))

bplot = sns.boxplot(data = df, y = '2017', x = '2019')

plt.suptitle('delitos por municipio en el a単o 2018',y=1.05)
```

```python
fig, axs = plt.subplots(figsize=(4,4))

bplot = sns.boxplot(data = df, y = '2018', x = '2020')

plt.suptitle('delitos por municipio en el a単o 2018',y=1.05)
```

```python
fig, axs = plt.subplots(figsize=(4,4))

bplot = sns.boxplot(data = df, y = '2019', x = 'Ago-2021')

plt.suptitle('delitos por municipio en el a単o 2018',y=1.05)
```

```python
myord1 = df['2017']
myord2 = df['2018']
myord3 = df['2019']
myord4 = df['2020']
myord5 = df['Ago-2021']
mycrossmap1 = pd.crosstab(myord1,myord2)
mycrossmap2 = pd.crosstab(myord2,myord3)
mycrossmap3 = pd.crosstab(myord3,myord4)
mycrossmap4 = pd.crosstab(myord4,myord5)
```

```python
sns.heatmap(mycrossmap1, annot=True, cmap='RdYlGn')
```

```python
sns.heatmap(mycrossmap2, annot=True, cmap='RdYlGn')
```

```python
sns.heatmap(mycrossmap3, annot=True, cmap='RdYlGn')
```

```python
sns.heatmap(mycrossmap4, annot=True, cmap='RdYlGn')
```
